<h2>Objective: Plot sin(x) in the range (0, 5pi)</h2>

```python
x = linspace(0, 5*pi, 100)
y = sin(x)
plot(x, y)
```

```python
#Remeber, plotting without axes labes or captions is a cardinal sin!
plot(x,y)
xlabel('x (radians)')
ylabel('sin(x)')
title('My first plot in Ipython Notebook!')
```
