<h3>objective 1: Load data</h3>

```python
def get_data(filename):
    #same as in 002: Lock and load but just returns the formatted data
    f = open(filename, 'rb')
    reader = csv.reader(f)
    headers = reader.next()
    data = []
    for line in reader:
        data.append(line)
    data = [[float(y) for y in x] for x in transpose(data)]
    #this time we 'return' the useful stuff without plotting
    return headers, data
```

```python
headers, data = get_data('MAPI.csv')
```

<h3>Objective 2: Plot the data using separate y-axes</h3>

```python
def twinplot(headers, data, leg=False):
    #It's useful to create a 'figure' object.
    #You can add as many lines or plots to this object as you like
    fig = figure()
    
    #add an axis to the plot
    ax = fig.add_subplot(111) #'111' means that your subplot takes up the whole page
    ln1, = ax.plot(data[0], data[1], color='blue', label=headers[1])
    ax.set_xlabel(headers[0])
    ax.set_ylabel(headers[1])

    ax2 = ax.twinx()
    ln2, = ax2.plot(data[0], data[2], color='red', label=headers[2])
    ax2.set_ylabel(headers[2])
    
    #you can also add a legend
    if leg:
        #by setting 'leg=False' as a function parameter the legend will be off by default
        ax.legend((ln1, ln2), (ln1.get_label(), ln2.get_label()))
    
    return fig
```

```python
plot = twinplot(headers, data)
```

```python
#To include the legend just set 'leg=True' when calling your function
plot = twinplot(headers, data, leg=True)
```
