```python
%pylab inline
import csv
```

```python
#new improved get_data function
def get_data(path, delim=','): #default delimeter is comma
    f = open(path, 'rb')
    
    #check for header text - True or False
    header = csv.Sniffer().has_header(f.read(1024))
    
    #you need to go back to the start of file
    f.seek(0)
    
    #create your reader object
    reader = csv.reader(f, delimiter=delim)
    
    header_text = None
    if header:
        #if headers exist get header text and skip row
        header_text = reader.next()
        reader.next()
    
    #get data
    data = []
    for line in reader:
        data.append(line)

    #transpose data for plotting        
    data = [[float(y) for y in x] for x in transpose(data)]
    
    #return headers and data as a single dictionary
    return {'headers': header_text, 'data': data}
```

```python
#get your data dictionary
d = get_data('data.csv')
```

```python
#you can check the contents of any dictionary 
d.keys()

d.get('headers', None)
```

```python
def scatterplot(d):
    
    headers = d.get('headers')
    data = d.get('data')

    #create your figure
    fig = figure()
    ax = fig.add_subplot(111)
    plt = ax.scatter(data[0], data[1], c=data[2], cmap='rainbow')
    cb = colorbar(plt)

    if headers:
        ax.set_xlabel(headers[0])
        ax.set_ylabel(headers[1])
        cb.set_label(headers[2])
        
    return fig, ax
```

```python
fig, ax = scatterplot(d)
```

```python
#linear regression in one line!
m, c = polyfit(d['data'][0], d['data'][1], 1)
print 'm = {0:.2f}, c = {1:.2f} eV'.format(m, c)
```

```python
#add line of best fit to fig
x = linspace(0, 1, 10) #nominal x data -- 10 data points between 0 and 1
y = m*x + c

#add plot to ax
ax.plot(x, y, '--', color='blue')
ax.set_xlim([0,1]) #define x-axis limits
ax.set_ylim([2.4, 3.6]) #define y-axis limits
```
